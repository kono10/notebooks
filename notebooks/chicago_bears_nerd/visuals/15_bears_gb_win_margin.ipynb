```python
import altair as alt
import pandas as pd
import numpy as np
import warnings 
import os
import glob

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_row', 500)
```

```python
file_path = "../data/team_ranks/*.csv"
```

```python

file_paths = glob.glob(os.path.join(file_path))
```

```python
pds = list()
```

```python
for file_path in file_paths:
    print(file_path)
    df = pd.read_csv(file_path).query("Rank <= 5")
    if df.Team.eq("Chicago").any():
        print("found data")
        pds.append(df)
```

```python
df_bears = pd.concat(pds)
```

```python
df_bears_only = df_bears[df_bears.Team.eq("Chicago") & df_bears["2022"] > 0]
```

```python
df = pd.read_csv("../data/team_ranks/_nfl_stat_incompletions_per_game_20220913.csv")
```

```python
chart_title = "Stat"
x_cfgs = {"shorthand":"Rank", "title":"Rank"}
y_cfgs = {"shorthand":"Team", "title":"Team"}

# background etc.
primary_color = "#0B162A" 
# font etc.
secondary_color = "#c83803"
axisLeftKwargs = dict(labelColor=secondary_color, titleColor=secondary_color, labelFontSize=15, titleFontSize=17)
axisBottomKwargs = dict(labelColor=secondary_color, titleColor=secondary_color,labelFontSize=15, titleFontSize=17)
legendKwargs = dict(labelColor=secondary_color, titleColor=secondary_color, strokeColor='gray', fillColor=primary_color, padding=10, cornerRadius=10,orient='top-right', labelFontSize=15, titleFontSize=15)
```

```python
chart = alt.Chart(df, width=1500, height=1500, title=chart_title)\
.mark_bar(color="white", opacity=1)\
.encode(
    x=alt.X(**x_cfgs,
    ),
    y=alt.Y(**y_cfgs,sort=alt.EncodingSortField(field="rank", order="descending")
    ),
    tooltip=[
        alt.Tooltip(shorthand="2022", title="Value"),
        alt.Tooltip(**x_cfgs),
        alt.Tooltip(**y_cfgs),
    ]
)



# put names inplace of bar
#chart.mark_text().encode(text="Team")\

chart\
.configure(background=primary_color)\
.configure_axis(gridColor="grey", gridDash=[6,4])\
.configure_axisLeft(**axisLeftKwargs)\
.configure_axisBottom(**axisBottomKwargs)\
.configure_title(color=secondary_color, fontSize=25)\
.configure_legend(**legendKwargs)
```
