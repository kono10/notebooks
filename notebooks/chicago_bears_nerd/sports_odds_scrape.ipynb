```python
for i, table in enumerate(tables[2:]):
    df = pd.read_html(str(table))
    df = df[0]
    df.columns = df.iloc[-2]
    df = df.iloc[:-2,: ]
    df["week"] = i + 1
    print(df)
    pd_list.append(df)
df = pd.concat(pd_list)
```

# Team Season Summary

```python
def find_tables(table_url):
    page = req.get(table_url, headers=headers).text
    soup = BeautifulSoup(page, "html.parser")
    return soup.find_all("table", attrs={"class":"soh1"})    
```

```python
url = "https://www.sportsoddshistory.com/nfl-game-team/?tm=CHI&d=2020#2022"
url_two = "https://www.sportsoddshistory.com/nfl-game-team/?tm=CHI&d=2010#2019"
url_three = "https://www.sportsoddshistory.com/nfl-game-team/?tm=CHI&d=2000#2009"
```

```python
tables = find_tables(url)[1:] + find_tables(url_two)[1:] + find_tables(url_three)[1:]
```

```python
pd_list = []
for table in tables:
    df = pd.read_html(str(table))
    df[0].columns = df[0].iloc[-1]
    df = df[0].iloc[:-1]
    df[["result_ats","spread"]] = df.Spread.str.split(" ", expand=True)
    df["result"] = df.Score.str.split(" ", expand=True)[0]
    df[["bears_pts","opp_pts"]] = df.Score.str.split(" ", expand=True)[1].str.split("-", expand=True)
    pd_list.append(df)

df = pd.concat(pd_list)
df.spread = df.spread.replace("PK", 0)
df.num_spread = pd.to_numeric(df.spread)
```

```python
df.opp_pts = pd.to_numeric(df.opp_pts)
df.bears_pts = pd.to_numeric(df.bears_pts)
```

```python
df["score_diff"] = df.bears_pts-df.opp_pts
```

```python
df["spread_margin"] = (df.bears_pts-df.opp_pts) + df.num_spread
```

```python
df_all = df[df.num_spread.gt(0)].sort_values("spread_margin", ascending=False).reset_index(drop=True)
```

```python
df.to_csv("data/2022_2000_chi_games_spread_history.csv")
```
