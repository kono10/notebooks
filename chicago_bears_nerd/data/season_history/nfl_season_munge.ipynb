```python
import pandas as pd
import pathlib
import os
```

```python
path = "/home/jkonovsky/development/notebooks/chicago_bears_nerd/data/season_history/"
```

```python
currentDir = pathlib.Path(path)
pd_list = []
```

```python
for file_path in currentDir.iterdir():
    if file_path.name.endswith("xls"):
        season_year = file_path.name.replace(".xls","")
        df = pd.read_html(file_path)[0]
        df["season_year"] = season_year
        pd_list.append(df)
df_all = pd.concat(pd_list)
```

```python
df = df_all[df_all.Date != "Playoffs"]
```

```python
output_path = os.path.join(path, "nfl_season_history.csv")
```

```python
df_all.to_csv(output_path)
```
