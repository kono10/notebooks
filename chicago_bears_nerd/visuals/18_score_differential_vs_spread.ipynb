```python
import altair as alt
import pandas as pd
import numpy as np
import warnings
```

```python
df = pd.read_csv("../data/2022_2000_chi_games_spread_history.csv")
```

```python
df.Date = pd.to_datetime(df.Date).dt.strftime("%Y-%m-%d")
```

```python
chart_title = "Bears Game Result vs Spread"
y_cfgs = {"shorthand":"spread", "title":"Spread"}
x_cfgs = {"shorthand":"score_diff", "title":"Score Diff (Bears Pts - Opponent Pts)"}
# background etc.
primary_color = "#0B162A" 
# font etc.
secondary_color = "#c83803"
axisLeftKwargs = dict(labelColor=secondary_color, titleColor=secondary_color, labelFontSize=15, titleFontSize=17)
axisBottomKwargs = dict(labelColor=secondary_color, titleColor=secondary_color,labelFontSize=15, titleFontSize=17)
legendKwargs = dict(labelColor=secondary_color, titleColor=secondary_color, strokeColor='gray', fillColor=primary_color, padding=10, cornerRadius=10,orient='top-right', labelFontSize=15, titleFontSize=15)
```

```python
# not working yet
#y_max, y_min = df[y_cfgs['shorthand']].max(), df[y_cfgs['shorthand']].min()
#x_max, x_min = df[x_cfgs['shorthand']].max(), df[x_cfgs['shorthand']].min()

chart = alt.Chart(df, width=800, height=800, title=chart_title)\
.mark_circle(size=500, color="white", opacity=.75)\
.encode(
    x=alt.X(**x_cfgs,
        scale=alt.Scale(domain=[-50, 50], domainMid=0)
    ),
    y=alt.Y(**y_cfgs,
                    scale=alt.Scale(domain=[-17, 17], domainMid=0)
    ),    
    tooltip=[
        "Opponent",
        "Date",
        "Week #",
        "Day",
        "Score",
        alt.Tooltip(**x_cfgs),
        alt.Tooltip(**y_cfgs),
        alt.Tooltip(shorthand="spread_margin", title="Spread - ScoreDiff")    ]
)


(chart)\
.configure(background=primary_color)\
.configure_axis(gridColor="grey", gridDash=[6,4])\
.configure_axisLeft(**axisLeftKwargs)\
.configure_axisBottom(**axisBottomKwargs)\
.configure_title(color=secondary_color, fontSize=25)\
.configure_legend(**legendKwargs)
```

```python
chart_title = "Bears Game Result vs Spread"
y_cfgs = {"shorthand":"spread_margin", "title":"Spread Beat"}

chart = alt.Chart(df.sort_values("Date"), width=1000, height=800, title=chart_title)\
.mark_bar(color="white", width=10)\
.encode(
    x=alt.X("Date",
    ),
    y=alt.Y(**y_cfgs,
    ),    
    tooltip=[
        "Opponent",
        "Date",
        "Week #",
        "Day",
        "Score",
        alt.Tooltip(**x_cfgs),
        alt.Tooltip(**y_cfgs),
        alt.Tooltip(shorthand="spread_margin", title="Spread - ScoreDiff")
    ]
)

(chart)\
.configure(background=primary_color)\
.configure_axis(gridColor="grey", gridDash=[6,4])\
.configure_axisLeft(**axisLeftKwargs)\
.configure_axisBottom(**axisBottomKwargs, labelOverlap=True, labelAngle=-45)\
.configure_title(color=secondary_color, fontSize=25)\
.configure_legend(**legendKwargs)
```
