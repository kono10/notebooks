```python
import altair as alt
import pandas as pd
import numpy as np
import warnings
```

```python
df_summary = pd.read_csv("../data/spread_history/chi_spread_summary.csv")
```

```python
df = pd.read_csv("../data/spread_history/nfl_2022_week_1_summary.csv")
```

```python

df_summary["W_ats_home_fav"] = pd.to_numeric(df_summary.home_fav_ats.str.split("-", expand=True)[0])

#df_summary["w_ats_home_dav"] = df_summary.home_fav_ats.str.split("-", expand=True)[0]

df_summary["L_ats_home_fav"] = pd.to_numeric(df_summary.home_fav_ats.str.split("-", expand=True)[1].str.split(" ", expand=True)[0])
```

```python
df_summary["win_percent"] = df_summary.W_ats_home_fav/(df_summary.W_ats_home_fav+ df_summary.L_ats_home_fav)
```

```python
df_summary
```

```python
df_summary[df_summary.year>20].win_percent.mean()
```

```python
df[["fav_result_ats","spread"]] = df.Spread.str.split(" ", expand=True)
df["fav_result"] = df.Score.str.split(" ", expand=True)[0]
df[["fav_pts","dog_pts"]] = df.Score.str.split(" ", expand=True)[1].str.split("-", expand=True)
df["num_spread"] = pd.to_numeric(df.spread)
df['score_diff'] = pd.to_numeric(df.fav_pts) - pd.to_numeric(df.dog_pts)
df['Is Bears'] = df.Favorite.eq("Chicago Bears") | df.Underdog.eq("Chicago Bears")
df['Total Spread Beat'] = df.score_diff + df.num_spread
```

```python
df
```

```python
chart_title = "Bears ATS as Home Favorites"
x_cfgs = {"shorthand":"year:O", "title":"Year"}
y_cfgs = {"shorthand":"win_percent", "title":"Bears ATS Win % as Home Favorites"}
# background etc.
primary_color = "#0B162A" 
# font etc.
secondary_color = "#c83803"
axisLeftKwargs = dict(labelColor=secondary_color, titleColor=secondary_color, labelFontSize=15, titleFontSize=17)
axisBottomKwargs = dict(labelColor=secondary_color, titleColor=secondary_color,labelFontSize=15, titleFontSize=17)
legendKwargs = dict(labelColor=secondary_color, titleColor=secondary_color, strokeColor='gray', fillColor=primary_color, padding=10, cornerRadius=10,orient='top-right', labelFontSize=15, titleFontSize=15)
```

```python
# not working yet
#y_max, y_min = df[y_cfgs['shorthand']].max(), df[y_cfgs['shorthand']].min()
#x_max, x_min = df[x_cfgs['shorthand']].max(), df[x_cfgs['shorthand']].min()

chart = alt.Chart(df_summary, width=1600, height=1000, title=chart_title)\
.mark_bar(color="white", opacity=1)\
.encode(
    x=alt.X(**x_cfgs,
        #scale=alt.Scale(domain=[-15, 15], domainMid=0)

    ),
    y=alt.Y(**y_cfgs,
        #scale=alt.Scale(domain=[-25, 25], domainMid=0)
    ),
    tooltip=[
        "year",
        "win_percent",
        alt.Tooltip(**x_cfgs),
        alt.Tooltip(**y_cfgs),
    ]
)
rule = alt.Chart(df_summary).mark_rule(color='red').encode(
    y='mean(win_percent):Q'
)

(chart + rule)\
.configure(background=primary_color)\
.configure_axis(gridColor="grey", gridDash=[6,4])\
.configure_axisLeft(**axisLeftKwargs)\
.configure_axisBottom(**axisBottomKwargs)\
.configure_title(color=secondary_color, fontSize=25)\
.configure_legend(**legendKwargs)
```
