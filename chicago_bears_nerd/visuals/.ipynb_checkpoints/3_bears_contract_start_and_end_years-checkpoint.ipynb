```python
import pandas as pd
```

```python
df = pd.read_csv("data/bears_2021_contract_info.csv")
df["contract_length"] = pd.to_numeric(df.contract_length)
df["contract_expires"] = df.contract_expires.replace("-", None)
df["contract_expires"] = pd.to_numeric(df.contract_expires)
```

```python
df["contract_start"] = df.contract_expires - df.contract_length
df['_contract_amount'] = df.contract_amount.map('${:,.0f}'.format)
```

```python
df
```

```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
from bokeh.sampledata.sprint import sprint
from bokeh.plotting import figure, show
from bokeh.layouts import gridplot
import bokeh.io

bokeh.io.reset_output()
bokeh.io.output_notebook()
```

```python
df = df[df.contract_amount > 1000000]
source = ColumnDataSource(df.sort_values("contract_amount", ascending=True))
tooltips = [
    ("Player Name", "@player_name"),
    ("Position", "@position"),
    ("Experience (yrs)", "@exp"),
    ("Contract Amount", "@_contract_amount"),
    ("Guaranteed Amount", "@guaranteed_amount"),
    ("Contract Start", "@contract_start"),
    ("Contract Expires", "@contract_expires"),
    ("Acquired", "@acquired"),
]

p = figure(width=1500, 
           height=2000, title="Chicago Bears Contract Start and End Summary Sorted by Contract Amount", 
           x_range=(df.contract_start.min()-1, df.contract_expires.max()+1), y_range=df.player_name.tolist()[::-1],
           tooltips=tooltips,
          )
p.hbar(y="player_name", left='contract_start', right='contract_expires', height=0.4, source=source, fill_color="orange", line_color="black")
p.title.text_font_size = "25px"
p.min_border_top = 75
p.min_border_bottom = 75 

show(p)
```

```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
from bokeh.sampledata.sprint import sprint

df = sprint.copy()  # since we are modifying sampledata

df.Year = df.Year.astype(str)
group = df.groupby('Year')
source = ColumnDataSource(group)

p = figure(y_range=group, x_range=(9.5,12.7), width=400, height=550, toolbar_location=None,
           title="Time Spreads for Sprint Medalists (by Year)")
p.hbar(y="Year", left='Time_min', right='Time_max', height=0.4, source=source)

p.ygrid.grid_line_color = None
p.xaxis.axis_label = "Time (seconds)"
p.outline_line_color = None

show(p)
```
