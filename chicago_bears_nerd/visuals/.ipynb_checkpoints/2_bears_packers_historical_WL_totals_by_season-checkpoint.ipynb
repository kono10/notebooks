```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import table


# configure pandas
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

_df = pd.read_csv("data/bears_packers_season_wl_record_history.csv", header=1)

df = _df[["Year", "Tm", "W", "Playoffs", "Lg"]]
df = df[df.Lg.eq("NFL")]
df["Tm"] = df["Tm"].str.replace("*","")
df["playoff_count"] = np.where(df.Playoffs.isnull(), 0, 1)
df = df.sort_values(["Tm", "Year"])
df = df.rename({"Chicago Staleys":"Chicago Bears"})
df["W_running_total"] = df.groupby("Tm")["W"].cumsum()
df["Playoffs_running_total"] = df.groupby("Tm")["playoff_count"].cumsum()
df = df.pivot_table(index="Year", columns="Tm", values=["W_running_total", "Playoffs_running_total"])

plt.style.use('dark_background')

fig, axes = plt.subplots(figsize=(20,15), nrows=2)
ax=axes[0]
ax2=axes[1]
#ax3=axes[2]

line_style = {"Chicago Bears":"Orange", "Green Bay Packers":"Green"}
df["W_running_total"].plot(ax=ax, style=line_style)
df["Playoffs_running_total"].plot(ax=ax2, style=line_style, legend=False)

ax.set_xlabel('', fontsize=15)
ax.set_ylabel('Running Total - Win Count', fontsize=15)
ax.set_title('Bears Packers Win Count Running Total Comparison', fontsize=25)

ax2.set_xlabel('Year', fontsize=15)
ax2.set_ylabel('Running Total - Playoff Count', fontsize=15)
ax2.set_title('Bears Packers Playoff Count Running Total Comparison', fontsize=25)

ax.grid(c="grey", linestyle=":")
ax2.grid(c="grey", linestyle=":")

#table(ax3, data=df.tail().T, loc="center")
ax3.axis("off")
#plt.figtext(.2, .3, df.iloc[-1].reset_index().to_string())

plt.show()
```
