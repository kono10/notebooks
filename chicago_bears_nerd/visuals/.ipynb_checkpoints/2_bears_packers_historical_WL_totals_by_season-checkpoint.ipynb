```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import numpy as np
from pandas.plotting import table


# configure pandas
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

_df = pd.read_csv("../data/bears_packers_season_wl_record_history.csv", header=1)
#_df = _df[_df.Year.gt(2007)]
```

```python
df = _df[["Year", "Tm", "W", "Playoffs", "Lg"]]
df = df[df.Lg.eq("NFL")]
df["Tm"] = df["Tm"].str.replace("*","")
df["playoff_count"] = np.where(df.Playoffs.isnull(), 0, 1)
df = df.sort_values(["Tm", "Year"])
df = df.rename({"Chicago Staleys":"Chicago Bears"})
df["W_running_total"] = df.groupby("Tm")["W"].cumsum()
df["Playoffs_running_total"] = df.groupby("Tm")["playoff_count"].cumsum()
df_wide = df
df = df.pivot_table(index="Year", columns="Tm", values=["W_running_total", "Playoffs_running_total"])
```

# Using Matplotlib

```python
#plt.style.use('dark_background')

#fig, axes = plt.subplots(figsize=(25,20), nrows=4)

fig = plt.figure(figsize=(25, 20))
gs = GridSpec(10, 1, wspace=0.25, hspace=0.25)    
ax=fig.add_subplot(gs[0:4])
ax2=fig.add_subplot(gs[4])
ax3=fig.add_subplot(gs[5:9])
ax4=fig.add_subplot(gs[9])


line_style = {"Chicago Bears":"Orange", "Green Bay Packers":"Green"}

# FIRST PLOT
df["W_running_total"].plot(ax=ax, style=line_style)
ax.set_xlabel('Year', fontsize=15)
ax.set_ylabel('Running Total - Win Count', fontsize=15)
ax.set_title('Bears Packers Win Count Running Total Comparison', fontsize=25)
ax.grid(c="grey", linestyle=":")

# DATA TABLE
# matplotlib table tutorial => https://towardsdatascience.com/simple-little-tables-with-matplotlib-9780ef5d0bc4
df_table = df_wide[df_wide.Year.gt(2000)].pivot_table(index="Tm", columns="Year", values="W_running_total") 
data = df_table.values
columns = df_table.columns
rows = df_table.index
the_table = ax2.table(cellText=data,
                      rowLabels=rows,
                      colLabels=columns,
                      loc='center')
the_table.scale(1, 1.5)

ax2.axis("off")

# SECOND PLOT
df["Playoffs_running_total"].plot(ax=ax3, style=line_style, legend=False)
ax3.set_xlabel('Year', fontsize=15)
ax3.set_ylabel('Running Total - Playoff Count', fontsize=15)
ax3.set_title('Bears Packers Playoff Count Running Total Comparison', fontsize=25)
ax3.grid(c="grey", linestyle=":")

# DATA TABLE
df_table = df_wide[df_wide.Year.gt(2000)].pivot_table(index="Tm", columns="Year", values="Playoffs_running_total") 
data = df_table.values
columns = df_table.columns
rows = df_table.index
the_table = ax4.table(cellText=data,
                      rowLabels=rows,
                      colLabels=columns,
                      loc='center')
the_table.scale(1, 1.5)

ax4.axis("off")

# Add a table at the bottom of the axes
#plt.figtext(0.95, 0.05, footer_text, horizontalalignment='right', 
plt.savefig("bears.png")

#plt.show()
```

# Using Altair

```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
df_wide.head()
```

```python
chart_title = "Total Wins All Time (starting 1922)"
y_cfgs = {"shorthand":"W_running_total", "title":"Wins Running Total"}
x_cfgs = {"shorthand":"Year", "title":"Year"}
# background etc.
primary_color = "#0B162A" 
# font etc.
secondary_color = "#c83803"
```

```python
df_wide.head()
```

```python
nearest = alt.selection(type='single', nearest=True, on='mouseover',fields=['Year'], empty='none')

line = alt.Chart(df_wide, width=1500, height=900, title=chart_title)\
.mark_line(size=3)\
.encode(
    x=alt.X(**x_cfgs),
    y=alt.Y(**y_cfgs),
    color=alt.Color("Tm", scale=alt.Scale(domain=["Chicago Bears", "Green Bay Packers"], range=[secondary_color, "white"])),

)

selectors = alt.Chart(df_wide).mark_point().encode(
    x='Year',
    opacity=alt.value(0),
).add_selection(
    nearest
)

points = line.mark_point().encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)

text = line.mark_text(align='left', fontSize=25).encode(
    text=alt.condition(nearest, 'W_running_total:Q', alt.value(' '))
)

rules = alt.Chart(df_wide).mark_rule(color='gray').encode(
    x='Year:Q',
).transform_filter(
    nearest
)
# Put the five layers into a chart and bind the data
alt.layer(
    line, selectors, points, rules, text
).configure(background=primary_color)\
.configure_axis(gridColor="grey", gridDash=[6,4])\
.configure_axisLeft(labelColor=secondary_color, titleColor=secondary_color, labelFontSize=15, titleFontSize=17)\
.configure_axisBottom(labelColor=secondary_color, titleColor=secondary_color,labelFontSize=15, titleFontSize=17)\
.configure_title(color=secondary_color, fontSize=25)\
.configure_legend(labelColor=secondary_color, titleColor=secondary_color, strokeColor='gray', fillColor=primary_color, padding=10,cornerRadius=10,orient='top-left', labelFontSize=15, titleFontSize=15)

```
